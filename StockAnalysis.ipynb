# Description of the Stock Movement

The purpose of this project is to analyse the stock that has been used further in a year and to analyse

Item Code: Unique code for a particular item

Item Desc : Description with size and additional information for a particular item.
Location: Location for the item iat the end of an accounting period in a ledger.

Item Type : Types of an item.

Brought Forward (BF): It represent the opening balance of an item at the start of a new accounting period or a new page in a ledger.

Goods Received Note (GRN): It represent the goods received at the end of an accounting period in a ledger

P/I	Purchase Invoice	Stock quantity recorded from a bill for items purchased.

C/P	Cash Purchase	Stock quantity received from a purchase paid for in cash.

SR	Stock Received	A general term for stock coming into the inventory.

ASM/I	Stock Assembly - In	Quantity of a finished good produced via assembly that is now in stock.

DA/I	(Likely) Delivery Advice - In	A document that accompanies incoming goods (similar to a delivery note).

ADJ/I	Stock Adjustment - In	Quantity added to stock due to an inventory count correction (e.g., finding previously unrecorded items).

XFER/I	Stock Transfer - In	Quantity of stock moved from another location or warehouse into this specific location.

UOM/I	Stock UOM Conversion - In	Quantity added due to a change in the Unit of Measurement (e.g., converting a box of 12 items into 12 single items).

C/N	Credit Note	A document used to process a return from a customer, adding the items back into stock.

D/R	Delivery Return	Items that were delivered to a customer but subsequently returned.-
